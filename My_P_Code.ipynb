```python
# Source: http://web.mta.info/developers/turnstile.html
#create data frame by download data sets from website then merge them

def get_data_parse_dt(week_nums):
    url = "http://web.mta.info/developers/data/nyct/turnstile/turnstile_{}.txt"
    dfs = []
    for week_num in week_nums:
        file_url = url.format(week_num)
        dfs.append(pd.read_csv(file_url))
    return pd.concat(dfs)
        
week_nums = [211002,210925,210918,210911,210904,210828,210821,210814,210807,210731,210724,210717,210710,210703]
MyDf= get_data_parse_dt(week_nums)
MyDf.head()
```

```python
#read & make query from DB file & tables 


from sqlalchemy import create_engine
engine = create_engine("sqlite:///Project_1.db")
DFA = pd.read_sql('SELECT * FROM Project_1_Table;', engine)
DFA.head()
```

```python
DFA.sample(20)
```

```python
# Take the date and time fields into a single datetime column


DFA["DATE_TIME"] = pd.to_datetime(DFA.DATE + " " + DFA.TIME, 
                                            format="%m/%d/%Y %H:%M:%S")
```

```python
DFA.sample(20)
```

```python
turnstiles_daily=DFA
```

```python
turnstiles_daily[["PREV_DATE", "PREV_ENTRIES"]] = (turnstiles_daily
                                                       .groupby(["C/A", "UNIT", "SCP", "STATION"])["DATE", "ENTRIES"]
                                                       .apply(lambda grp: grp.shift(1)))
```

```python
turnstiles_daily.sample(10)
```

```python
#ADD daily_entries


turnstiles_daily ['DAILY_ENTRIES'] = turnstiles_daily['ENTRIES'] -  turnstiles_daily['PREV_ENTRIES']
turnstiles_daily.head()
```

```python
#cleanning diff & apply
#Create methode that remove -tive values & outlier values(have been reset)


def get_daily_counts(row, max_counter):
    counter = abs(row["ENTRIES"] - row["PREV_ENTRIES"])
    if counter > max_counter:
        counter = row["ENTRIES"]
    return counter

#crete diff_ & apply above method 


turnstiles_daily['diff_'] = turnstiles_daily.apply(get_daily_counts, axis=1, max_counter=40000)
```

```python
#
turnstiles_daily = turnstiles_daily[~((turnstiles_daily['DAILY_ENTRIES']>0)
                   & (turnstiles_daily['DAILY_ENTRIES']!=turnstiles_daily['diff_']))]
```

```python
#Drop null row

turnstiles_daily.dropna(inplace=True)
turnstiles_daily.head()

turnstiles_daily
```

```python
#AM & PM col


import numpy as np
turnstiles_daily["TIME_INTERVAL"] = (
    np.where( (turnstiles_daily.DATE_TIME.dt.hour < 12), "AM","PM"))
```

```python
turnstiles_daily
```

```python
#create days cols


turnstiles_daily['DAY_NUM']=pd.to_datetime(turnstiles_daily['DATE']).dt.weekday
turnstiles_daily['DAY']=pd.to_datetime(turnstiles_daily['DATE']).dt.day_name()
```

```python
#create week day -end col


turnstiles_daily["WEEK"] = (
    np.where( (turnstiles_daily.DATE_TIME.dt.weekday <= 4), "Weekdays"
         , "Weekends"
         ))
```
